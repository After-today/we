* subplot

fig,add_subplot(a,b,c)

a,b表示将fig分割成a*b的区域

c表示当前选中要操作的区域 注意：从1开始编号

```python
#ax1 = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,1)#表示分割成四块区域 但jupyter有个bug图上不会显示分开的 
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
# 在subplot上作图
import numpy as np

random_arr = np.random.randn(100)
#print random_arr

# 默认是在最后一次使用subplot的位置上作图，但是在jupyter里无效
plt.plot(random_arr)
plt.show()
```

```python
# 在指定subplot作图

import scipy as sp
from scipy import stats

x = np.linspace(-5, 15, 50)
#print x.shape

# 绘制高斯分布
plt.plot(x, sp.stats.norm.pdf(x=x, loc=5, scale=2))

# 叠加直方图
plt.hist(sp.stats.norm.rvs(loc=5, scale=2, size=200), bins=50, normed=True, color='red', alpha=0.5)
plt.show()
```

```python
# 绘制直方图
#plt.hist(np.random.randn(100), bins=10, color='b', alpha=0.3)bins表示最后分为十组 alpha表示透明度
#plt.show()
plt.hist(np.random.randn(100), bins=10,color='yellow',alpha=0.4)
```

```python
# 绘制散点图
#x = np.arange(50)
#y = x + 5 * np.random.rand(50)
#plt.scatter(x, y)
x = np.arange(50)
y = x+ np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
# 柱状图
x = np.arange(5)#一位数组有五个类别 要做柱状图
y1, y2 = np.random.randint(1, 25, size=(2, 5))#两位五列的随机数
width = 0.25#指定数据的宽度
ax = plt.subplot(1,1,1)
ax.bar(x, y1, width, color='r')
ax.bar(x+width, y2, width, color='g')#x的位置加一个位移
ax.set_xticks(x+width)#设置在什么地方进行输出
ax.set_xticklabels(['a', 'b', 'c', 'd', 'e'])#标签
plt.show()
```

```python
# 矩阵绘图 经常用于表示三个维度的关系
#m = np.random.rand(10,10)生成十行十列的矩阵
#print m
#plt.imshow(m, interpolation='nearest', cmap=plt.cm.ocean)
#plt.colorbar()
#plt.show()
m = np.random.rand(10,10)
print(m)
plt.imshow(m,interpolation = 'nearest',cmap = plt.cm.ocean)#cmap要使用哪种配色方案
plt.colorbar()#颜色的区域
plt.show()
```

* plt.subplots()

同时返回创建的figure和subplot对象数组

推荐使用这种方式来创建多图表

```python
 #fig, subplot_arr = plt.subplots(2,2)#可以用此来创建多个图表plt subplot返回值有两个，一个fig是figur对象，一个是数组 也就是我最后2×2的数组
#subplot_arr[0,0].hist(np.random.randn(100), bins=10, color='b', alpha=0.3)#我们想要在0行0列有一个直方图 其他默认都没有 就是空白的
#plt.show()
fig, subplot_arr = plt.subplots(2,2)
subplot_arr[0,0].hist(np.random.randn(100),bins=10,color='pink',alpha=0.3)
subplot_arr[0,1].scatter(np.random.randn(100),np.random.randn(100)+1)
plt.show()
```
